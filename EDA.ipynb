```python
'''
Q-11>  Identify and remove outliers in the km_driven column using the IQR method ?
Sol:
'''
import pandas as pd
path=r'C:\Users\hp\OneDrive\Desktop\BIKE DETAILS.csv'
df=pd.read_csv(path)
Q1=df['km_driven'].quantile(0.9)
Q3=df['km_driven'].quantile(0.60)
IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
filtered_df=df[(df['km_driven']>=lower_bound)&(df['km_driven']<=upper_bound)]
removed_outliers_count=len(df)-len(filtered_df)
print(f"Number of outliers removed: {removed_outliers_count}")


```

```python
'''
Q-12> Perform a bivariate analysis to visualize the relationship between year and selling_price ?
Sol:
'''
import pandas as pd
import matplotlib.pyplot as plt
path = r'C:\Users\hp\OneDrive\Desktop\BIKE DETAILS.csv'
df = pd.read_csv(path)
bins = [0,50000,100000,150000,200000,float('inf')]  
labels = ['0-50k','50k-100k','100k-150k','150k-200k','200k+']
df['price_category']=pd.cut(df['selling_price'],bins=bins,labels=labels)
price_by_year = df.groupby(['year', 'price_category'], observed=True).size().unstack()
for year in price_by_year.index:
    plt.figure(figsize=(5,5))
    price_by_year.loc[year].plot(kind='pie',autopct='%1.1f%%',colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0'])
    plt.title(f'Price Distribution for Year {year}')
    plt.ylabel('')  # Hide ylabel
    plt.tight_layout()
    plt.show()
```

```python
'''
Q-13> What is the average depreciation in selling price based on the bike's age (current year - manufacturing year) ?
Sol:
'''
import pandas as pd
import matplotlib.pyplot as plt
path=r'C:\Users\hp\OneDrive\Desktop\BIKE DETAILS.csv'
df=pd.read_csv(path)
current_year=2025
df['bike_age']=current_year - df['year']
df['depreciation']=df['ex_showroom_price'] - df['selling_price']
avg_depreciation_by_age=df.groupby('bike_age')['depreciation'].mean()
print("Average Depreciation by Bike Age:")
print(avg_depreciation_by_age)
plt.figure(figsize=(19,7))
avg_depreciation_by_age.plot(kind='line', color='green', marker='o')
plt.title('Average Depreciation in Selling Price Based on Bike Age')
plt.xlabel('Bike Age (Years)')
plt.ylabel('Average Depreciation (INR)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
'''
Q-14> Which bike names are priced significantly above the average price for their manufacturing year ?
Sol:
'''
import pandas as pd
path=r'C:\Users\hp\OneDrive\Desktop\BIKE DETAILS.csv'
df=pd.read_csv(path)
avg_price_by_year=df.groupby('year')['selling_price'].mean()
threshold_multiplier=12.9
df['significantly_above_avg']=df.apply(lambda row: row['selling_price']>avg_price_by_year[row['year']]* threshold_multiplier,axis=1)
bikes_above_avg = df[df['significantly_above_avg']]
significantly_above_avg_bikes=bikes_above_avg['name']
print("---------------------Bikes priced significantly above the average price for their manufacturing year--------------------")
print("\n")
print(significantly_above_avg_bikes)
```

Description:

The dataset contains details about used cars listed for sale, such as brand, model, selling price, kilometers
driven, fuel type, and transmission type. This information is valuable for predictive modeling, market analysis,
and understanding customer preferences

Features:

Car_id: A unique identifier for each car in the dataset, helping to track individual car entries
Date: The date when the car sale transaction took place, formatted as YYYY-MM-DD
Customer Name: The name of the customer who purchased the car, represented as a string
Gender: The gender of the customer, categorized as "Male" or "Female.
Annual Income: The customer's annual income in US dollars, represented as a numeric value
Dealer_Name: The name of the dealership selling the car, represented as a string
Company: The manufacturer or brand name of the car, such as "Toyota," "Ford," etc
Model: The specific model name of the car, such as "Corolla," "Civic," etc
Engine: The engine type of the car, such as "V6," "I4," etc
Transmission: The type of transmission in the car, either "Manual" or "Automatic.
Color: The color of the car, represented as a string (e.g., "Red," "Blue")
Price ($): The selling price of the car in US dollars
Dealer_No: A unique identifier for each car dealer in the dataset
Body Style: The body style of the car, such as "Sedan," "SUV," etc
Phone: The phone number of the customer who purchased the car
Dealer_Region: The geographical region of the car dealer, such as "North," "South," etc

```python
'''
Q-12> Is there any seasonal trend in car sales based on the date of sale?
Sol:
'''
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
path = r'C:\Users\hp\OneDrive\Desktop\Car Sale.csv'
df = pd.read_csv(path)
print("Missing values in Date:", df['Date'].isnull().sum())
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y')
df['Month'] = df['Date'].dt.month
df['Month_Name'] = df['Date'].dt.strftime('%b')
sales_by_month = df.groupby(['Month', 'Month_Name'])['Date'].count().reset_index()
sales_by_month = sales_by_month.rename(columns={'Date': 'Sales Count'})
sales_by_month = sales_by_month.sort_values('Month')  # Ensure months are in calendar order
print("\nCar Sales Count by Month:")
print(sales_by_month[['Month_Name', 'Sales Count']])
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'
df['Season'] = df['Month'].apply(get_season)
sales_by_season = df.groupby('Season')['Date'].count().reset_index()
sales_by_season = sales_by_season.rename(columns={'Date': 'Sales Count'})
season_order = ['Winter', 'Spring', 'Summer', 'Fall']
sales_by_season = sales_by_season.set_index('Season').reindex(season_order).reset_index()
print("\nCar Sales Count by Season:")
print(sales_by_season)
plt.figure(figsize=(10, 6))
sns.barplot(x='Month_Name', y='Sales Count', data=sales_by_month)
plt.title('Car Sales by Month', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Sales Count', fontsize=12)
plt.tight_layout()
plt.show()
plt.figure(figsize=(8, 6))
sns.barplot(x='Season', y='Sales Count', data=sales_by_season, order=season_order)
plt.title('Car Sales by Season', fontsize=14)
plt.xlabel('Season', fontsize=12)
plt.ylabel('Sales Count', fontsize=12)
plt.tight_layout()
plt.show()
```
